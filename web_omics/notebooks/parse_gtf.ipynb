```python
import glob
import sys
sys.path.append('../linker')

from GTF import lines, dataframe, parse
```

```python
gene_names = {}
for filename in sorted(glob.glob("/Users/joewandy/Downloads/gtf_to_parse/*.gtf.gz")):

    collision = 0
    try:
        gtf_dict = lines(filename)
        for d in gtf_dict:
            try:
                gene_id = d['gene_id']
                gene_name = d['gene_name']
                gene_names[gene_id] = gene_name
            except KeyError:
                continue
    except IndexError:
        continue
    finally:
        print(filename, len(gene_names))                
```

```python
import pickle
pickle.dump(gene_names, open('../static/data/gene_names.p', 'wb'))
```
